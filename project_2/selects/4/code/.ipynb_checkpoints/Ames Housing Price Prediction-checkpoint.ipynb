## 1. Methodology

Data cleaning is necessary in preparing the data for analysis and subsequent modeling. The approach taken on our training and test set is as follows:

Converting all the column names to **standardised lower case** and **snakecase convention**

**Imputed missing and null values**: as the dataset is not huge, it is not advisable to drop rows or columns with missing data to prevent data leakage. Hence, it would make better sense to fill in missing and null values by using logical deductions based on the intepretations from the data dictionary. For example, when imputing 330 missing values in lot frontage by filling them with their neighborhood's median lot frontage value. When it came to imputing missing data on the test set, we used the mode to fill in the missing value for 'electrical'. 

**Fixed data errors**: As the a property where Garage Yr Blt = 2207, looked like a data entry error, we fixed this by replacing the garage year built to the year the house was built.

**Box-plots**: Reviewing box-plots for all categorical, nominal data and dropping attributes with dominating class or ones with similar median and little variance across classes.

**Correlation heatmap, Scatter plot, Multicollinearity Analysis**: Heatmap for numerical variables is constructed with low contributing variables dropped. Scatterplot is review to understand the distribution of variables and outliers are addressed.
Multicollinearity analysis is also performed and for variables with high collinearity, the one with higher correlation is retained and the other dropped.

**Mapped ordinal variables to numerical values**: In the case of an ordinal variable like garage condition, we mapped the following: {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'NA': 0}, with NA ie no garage being 0 and Excellent being a value of 5.

**Feature engineering**: Variables are relooked at and some of the classes within the same variables are collapse for better data management. For instance, for MS zoning, we have recategorised 8 classes into 2 by mapping the data into residential (R) and non-residential (NR)

**One-hot encoding for categorical nominal variables**: These variables were converted into series of binary variables that represent whether a particular feature was present in a given property or not. 

**Train-Test Split and Scaling**: The train set was train-test split for validation and scaled before processing by regression models.

**Regression Models**: Train set was processed by 3 models, MLR, Ridge CV and LassoCV and the model with best R2 and RMSE results is selected as final model for testing

**Training of selected model & prediction**:Training model is trained and using test(holdout set) for prediction of housing price

### A. Categorical Variables

Let us first examine the categorical data that by going through the data dictionary to identify wich ones can be converted into numerical format.
There are altogether 42 categorical variables.
Below is a list of columns that we can transform based on the ordinal features names to numerical replacement range.

| Category | Ordinal Features Names | Categorical Order | Numerical Replacement Range | 
|------|------------------------|------------------------------|-----------------------|
| 1 | exter_qual | Po -> Fa -> TA -> Gd -> Ex | 1 - 5 | 
| 1 | exter_cond | Po -> Fa -> TA -> Gd -> Ex | 1 - 5 | 
| 1 | heating_qc | Po -> Fa -> TA -> Gd -> Ex | 1 - 5 | 
| 1 | kitchen_qual | Po -> Fa -> TA -> Gd -> Ex | 1 - 5 | 
| 2 | bsmt_qual | NA -> Po -> Fa -> TA -> Gd -> Ex | 0 - 5 | 
| 2 | bsmt_cond | NA -> Po -> Fa -> TA -> Gd -> Ex | 0 - 5 | 
| 2 | fireplace_qu | NA -> Po -> Fa -> TA -> Gd -> Ex | 0 - 5 | 
| 2 | garage_qual | NA -> Po -> Fa -> TA -> Gd -> Ex | 0 - 5 | 
| 2 | garage_cond | NA -> Po -> Fa -> TA -> Gd -> Ex | 0 - 5 
| 3 | bsmt_fin_type1 | NA -> Unf -> LwQ -> Rec -> BLQ -> ALQ -> GLQ | 0 - 6|
| 3 | bsmt_fin_type2 | NA -> Unf -> LwQ -> Rec -> BLQ -> ALQ -> GLQ | 0 - 6| 
| 4 | garage_finish | NA -> Unf -> RFn -> Fin | 0 - 3 | 
| 5 | pool_qc | NA -> Fa -> TA -> Gd -> Ex | 0 - 4 | 
| 6 | bsmt_exposure | NA -> No -> Mn -> Av -> Gd | 0 - 4 | 
| 7 | fence | NA -> MnWw -> GdWo -> MnPrv -> GdPrv | 0 - 4 | 

```python
# create a dictionary for category 1 quality scores
scores = {'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1}

# change the values for variables in category 1 for train dataset
train['exter_qual'] = train['exter_qual'].map(scores)
train['exter_cond'] = train['exter_cond'].map(scores)
train['heating_qc'] = train['heating_qc'].map(scores)
train['kitchen_qual'] = train['kitchen_qual'].map(scores)

# Sanity Check
train[['exter_qual','exter_cond','heating_qc','kitchen_qual']].head(1)
```
